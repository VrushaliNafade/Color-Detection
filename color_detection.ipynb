```python
import cv2
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
 image = cv2.imread('/content/colorful-flowers-garden-macro.jpg')
```

```python
plt.figure(figsize = (20,8))
plt.imshow(image)
```

```python
grid_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize = (20,8))
plt.imshow(grid_RGB)
```

```python
#Detecting Yellow color.
grid_HSV = cv2.cvtColor(grid_RGB, cv2.COLOR_RGB2HSV)
```

```python
lower = np.array([25,150,50])
upper = np.array([35,255,255])
```

```python
mask = cv2.inRange(grid_HSV, lower, upper)
```

```python
plt.figure(figsize = (20,8))
plt.imshow(mask)
```

```python
result = cv2.bitwise_and(grid_RGB, grid_RGB, mask=mask)
```

```python
plt.figure(figsize = (20,8))
plt.imshow(result)
```

**Why do we use the HSV color space in computer vision or image processing?**

Answer - The reason we use HSV colorspace for color detection/thresholding over RGB/BGR is that HSV is more robust towards external lighting changes. This means that in cases of minor changes in external lighting (such as pale shadows,etc. ) Hue values vary relatively lesser than RGB values.
